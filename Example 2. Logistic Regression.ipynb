```python
from train import train
from neural_networks import full_forward_propagation, get_accuracy_value

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

random_state=42

#DATASET-------------------------------
data = pd.read_csv('Datasets/iris.data', header = None)
data.loc[:,4] = pd.Categorical(data.loc[:,4])

data_X = data.loc[:, 0:3]
data_Y = pd.get_dummies(data.loc[:,4], prefix = 'category')

data_X = data_X.values
data_Y = data_Y.values

#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, test_size=0.1, random_state=42)

#NEURAL NETWORK ARCHITECTURE------------
NN_ARCHITECTURE = [
    {"input_dim": 4, "output_dim": 3, "activation": "softmax"}
]

#HYPERPARAMETERS-----------------------
N_nn = 20

epochs = 500
learning_rate = 0.1
var_epsilon = 0

method = "nogradient"
cost_type = 'cross_entropy_T' 
acc_type = 'f1_score'

verbose = False

batch_size = X_train.shape[0]
n_batches  = 1
```

## Derivative Free -> kernel_a = 0.01, gamma1 = 0, gamma2 = 0

```python
kernel_a = 0.001
alpha_init = 0
alpha_rate = 0
beta = 0
gamma = 0


#Trainining----------------
params, mean_param = train(np.transpose(X_train), np.transpose(y_train), NN_ARCHITECTURE, epochs, learning_rate, method, n_batches, batch_size, cost_type, N_nn, kernel_a, alpha_init, alpha_rate, beta, gamma, verbose, var_epsilon)

#TEST SET PREDICTION AND ACCURACY OF THE FIRST PARTICLE
Y_test_hat, _ = full_forward_propagation(np.transpose(X_test), params[0], NN_ARCHITECTURE)
acc_test = get_accuracy_value(Y_test_hat.T, y_test, acc_type)
print("Test set accuracy measure using params[0]: {:.2f}".format(acc_test))

#TEST SET PREDICTION AND ACCURACY OF MEAN TENSOR OF NETWORK
Y_test_hat, _ = full_forward_propagation(np.transpose(X_test), mean_param, NN_ARCHITECTURE)
acc_test = get_accuracy_value(Y_test_hat.T, y_test, acc_type)
print("Test set accuracy measure using mean_param: : {:.5f}".format(acc_test))
```
